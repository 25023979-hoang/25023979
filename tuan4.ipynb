```python
numbers = []
while True:
    n = int(input("Nhập số(nhập -1 để dừng): "))
    if n == -1: break
    numbers.append(n)
if numbers:
    print("Max:", max(numbers))
    print("Min:", min(numbers))
else:
    print("Danh sách trống")
```

bai 12

```python
s = input("Nhập chuỗi: ")
dem = 0
i = 0
while True:
    try:
        if s[i] != " " and (i == 0 or s[i-1] == " "):
            dem += 1
        i += 1
    except IndexError:
        break
print("Số từ là:", dem)
```

```python
s = input("Nhập chuỗi: ")
tu_dau_tien = ""
for i in range(len(s)):
    if s[i] != " ":
        tu_dau_tien += s[i]
    elif tu_dau_tien != "":
        break

print("Từ đầu tiên là:", tu_dau_tien)
```

bai 14

```python
s = input("Nhập chuỗi: ")
ketqua = ""
for kytu in s:
    if kytu.islower():
        ketqua += kytu.upper()
    elif kytu.isupper():
        ketqua += kytu.lower()
    else:
        ketqua += kytu
print("Chuỗi sau khi đổi:", ketqua)
```

```python
s = input()
so=0
inthuong=0
inhoa=0
for kytu in s:
    if kytu.islower():
        inthuong += 1
    elif kytu.isupper():
        inhoa += 1
    elif kytu.isdigit():
        so += 1
    else:
        pass
print(f'so: {so}, chu cai thuong: {inthuong}, chu cai hoa: {inhoa}')
```
