```python
def is_perfect_number(n):
    if n <= 0: 
        return False
    
    tong_uoc = 0
    for i in range(1, n):
        if n % i == 0:
            tong_uoc += i
    return tong_uoc == n
num = int(input())

if is_perfect_number(num):
    print(f"{num} là số hoàn hảo.")
else:
    print(f"{num} không phải là số hoàn hảo.")
```

```python
def check_dc(s1, s2):
    if len(s1) != len(s2):
        return False
    
    d1 = {}
    d2 = {}
    
    for c1, c2 in zip(s1, s2):
        if c1 in d1:
            if d1[c1] != c2:
                return False
        else:
            d1[c1] = c2
            
        if c2 in d2:
            if d2[c2] != c1:
                return False
        else:
            d2[c2] = c1
            
    return True

t1 = input("Nhap tu 1: ")
t2 = input("Nhap tu 2: ")
print(check_dc(t1, t2))
```

```python
import math

def tinh_goc(x1, y1, x2, y2, x3, y3):
    ba = (x1-x2, y1-y2)
    bc = (x3-x2, y3-y2)
    tu = ba[0]*bc[0] + ba[1]*bc[1]
    mau = math.sqrt(ba[0]**2 + ba[1]**2) * math.sqrt(bc[0]**2 + bc[1]**2)
    return math.degrees(math.acos(tu / mau))

print(tinh_goc(0,1, 0,0, 1,0)) [cite: 4]
```

```python
def sum_unq(n):
    if n <= 0: return 0
    t = 0
    for i in range(1, n + 1):
        s = str(i)
        if len(set(s)) == len(s):
            t += i
    return t

n = int(input("Nhap n: "))
print(sum_unq(n))
```
