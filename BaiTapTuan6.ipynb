```python
def tao_dic(dong):
    dic = {}
    ds_cap = dong.split()
    for cap in ds_cap:
        k, v = cap.split(':')
        if k not in dic:
            dic[k] = []
        dic[k].append(v)
    return dic

nhap = input()
print(tao_dic(nhap))
```

```python
def dem_so(ds):
    kq = {'positives': 0, 'negatives': 0, 'zeros': 0}
    for x in ds:
        if x > 0: 
            kq['positives'] += 1
        elif x < 0: 
            kq['negatives'] += 1
        else: 
            kq['zeros'] += 1
    return kq

ds = list(map(int, input().split()))
print(dem_so(ds))
```

```python
def tim_cap(ds, k):
    kq = []
    n = len(ds)
    for i in range(n):
        for j in range(i + 1, n):
            if ds[i] + ds[j] == k:
                kq.append((ds[i], ds[j]))
    return sorted(kq)

ds = list(map(int, input().split()))
k = int(input())
print(tim_cap(ds, k))
```

```python
def tim_max(ds):
    dem = {}
    for x in ds:
        dem[x] = dem.get(x, 0) + 1
    max_lan = max(dem.values())
    ds_max = [k for k, v in dem.items() if v == max_lan]
    return min(ds_max)

ds = list(map(int, input().split()))
print(tim_max(ds))
```

```python
def vt_max(ds):
    if not ds: return -1
    m = ds[0]
    p = 0
    for i in range(len(ds)):
        if ds[i] > m:
            m = ds[i]
            p = i
    return p

a = list(map(int, input().split()))
print(vt_max(a))
```

```python
def day_zero(ds):
    khac = []
    khong = []
    for x in ds:
        if x == 0:
            khong.append(0)
        else:
            khac.append(x)
    return khac + khong

a = list(map(int, input().split()))
print(day_zero(a))
```

```python
def tong_cd(ds):
    tong = 0
    for i in range(1, len(ds) - 1):
        if ds[i] > ds[i-1] and ds[i] > ds[i+1]:
            tong += ds[i]
    return tong

a = list(map(int, input().split()))
print(tong_cd(a))
```

```python
def kt_day(ds):
    tang = True
    giam = True
    for i in range(len(ds) - 1):
        if ds[i] >= ds[i+1]: tang = False
        if ds[i] <= ds[i+1]: giam = False
    
    if tang: return "Tang thuc su"
    if giam: return "Giam thuc su"
    return "Vo danh"

a = list(map(int, input().split()))
print(kt_day(a))
```
