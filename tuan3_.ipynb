```python
import math
while True:
    try:
        n = int(input("Nhập một số nguyên dương: "))
        if n > 0:
            break
        else:
            print("Vui lòng nhập số lớn hơn 0!")
    except ValueError:
        print("Đầu vào không hợp lệ. Vui lòng nhập một số nguyên!")
is_prime = True
if n < 2:
    is_prime = False
else:
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            is_prime = False
            break
if is_prime:
    print(f"{n} là số nguyên tố.")
else:
    print(f"{n} không phải là số nguyên tố.")
```

```python
import math
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True
try:
    while True:
        line = input('nhap a va b theo dang x y: ')
        if not line:
            break
        data = list(map(int, line.split()))
        if len(data) < 2:
            break
        a, b = data
        tong_snt = 0
        for num in range(a, b + 1):
            if is_prime(num):
                tong_snt += num
        print(tong_snt)
except EOFError:
    pass
```

```python
def palindrome(num):
    s = str(num)
    return s == s[::-1]
def daonguoc(num):
    s = str(num)
    return int(s[::-1])
n = int(input('nhap so n: '))
buoc = 0
while not palindrome(n):
    n = n + daonguoc(n)
    buoc += 1
print(f'so buoc: {buoc}\ngia tri n: {n}')
```

```python
import math
n = int(input('nhap so nguyen duong n: '))
ket_qua = []
for i in range(int(math.sqrt(n)) + 1):
    so_cp = i * i
    if so_cp > n:
        break  
    s = str(so_cp)
    if len(set(s)) == len(s):
        ket_qua.append(s)
if not ket_qua:
    print("(no number)")
else:
    print(" ".join(ket_qua))
```

```python
def dodai(m):
    count = 1
    while m > 1:
        if m % 2 == 0:
            m //= 2
        else:
            m = 3 * m + 1
        count += 1
    return count
n = int(input())
max_len = 0
best_x = 1
for i in range(1, n + 1):
    current_len = dodai(i)
    if current_len > max_len:
        max_len = current_len
        best_x = i
print(best_x, max_len)
```

```python
import math
def tinh_tong_uoc(n):
    if n < 2:
        return 0
    tong = 1
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            tong += i
            if i != n // i:
                tong += n // i
    return tong
a = int(input())
b = int(input())
if tinh_tong_uoc(a) == b and tinh_tong_uoc(b) == a:
    print("true")
else:
    print("false")
```
